```python
data_map = data.loc[(data.Lat>40)&(data.Long<=-5)]   # лонг - у : лат - х 
data1=data[('OFFENSE_CODE_GROUP')].value_counts().reset_index()
data_bad = data1['OFFENSE_CODE_GROUP'].value_counts() 
data1.head(9) 
col=['Pastel2', 'Pastel1', 'YlOrBr_r', 'PuBu', 'autumn', 'BuGn_r', 'BuPu', 'Set2', 'CMRmap']
x=[] 
for n in range(9):
  g=data_map[data_map['OFFENSE_CODE_GROUP']==data1['index'][n]]
  x.append(g)
def group(i):  
  return x[i] 
def graf(x, i ,j): 
  sns.scatterplot(data=group(x), x='Lat', y='Long' , hue ='OFFENSE_CODE_GROUP', size=1, ax =ax[i][j], palette=col[x], legend='full') 
f, ax=plt.subplots(ncols=3, nrows=3, figsize = (20,24)) 
i = 0 
j = 0 
for r in range(9): 
  graf(r, i ,j) 
  if i == 2: 
    j+=1 
    i=0 
  else: 
    i+=1 
```
